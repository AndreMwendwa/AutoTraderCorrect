```python
data_2019_23 = pd.read_csv('izev_data\\izev-webstats_fy-2019-20-to-fy-2022-23_en_updated-november-2025.csv')
data_2019_23
```

```python
dfSummary(data_2019_23)
```

Things I have to do: <br><ul><li>Filter out leases <li>Filter out non-Teslas <li>Filter out NA rows (for 2019-2023) <li>Drop last 3 characters and space from postal code. <li>Filter out model years not 2021-2024

2023 - 24 Data First

```python
purchase_condition = (data_2023_24['Purchase or Lease'] == 'Purchase')
tesla_condition = (data_2023_24['Vehicle Make'] == 'Tesla')
model_year_condition = ((data_2023_24['Vehicle Year'] >= 2021) & (data_2023_24['Vehicle Year'] <= 2024))
period_condition = (data_2023_24['Incentive Request Date'] >= '2021-01-01') & (data_2023_24['Incentive Request Date'] <= '2024-03-31')
na_condition = data_2023_24['Vehicle Model'].notna()
    
data_2023_24_filtered = data_2023_24[purchase_condition & tesla_condition & model_year_condition & period_condition & na_condition].reset_index(drop=True)
data_2023_24_filtered
```

```python
data_2023_24_filtered['Battery, Plug-in Hybrid, or Fuel Cell EV (BEV, PHEV or FCEV)'].value_counts()
```

Passes sanity check: All are BEVs

2019 - 2023 Data

```python
purchase_condition = (data_2019_23['Purchase or Lease'] == 'Purchase')
tesla_condition = (data_2019_23['Vehicle Make'] == 'Tesla')
model_year_condition = ((data_2019_23['Vehicle Year'] >= 2021) & (data_2019_23['Vehicle Year'] <= 2024))
period_condition = (data_2019_23['Incentive Request Date'] >= '2021-01-01') & (data_2019_23['Incentive Request Date'] <= '2024-03-31')
na_condition = data_2019_23['Vehicle Model'].notna()

data_2019_23_filtered = data_2019_23[purchase_condition & tesla_condition & model_year_condition & period_condition & na_condition].reset_index(drop=True)
data_2019_23_filtered
```

```python
data_2019_23_filtered['Battery, Plug-in Hybrid, or Fuel Cell EV (BEV, PHEV or FCEV)'].value_counts()
```

Passes Sanity check: all BEVs

Removing last 3 characters of postal code

```python
# Remove last 3 characters of postal code
data_2023_24_filtered['Dealership Postal Code'] = data_2023_24_filtered['Dealership Postal Code'].str[:-3].str.strip()
data_2023_24_filtered['Dealership Postal Code'].value_counts()
```

```python
# Remove last 3 characters of postal code
data_2019_23_filtered['Dealership Postal Code'] = data_2019_23_filtered['Dealership Postal Code'].str[:-3].str.strip()
data_2019_23_filtered['Dealership Postal Code'].value_counts()
```

So it turns out that Teslas are all sold from some fictional postal code. So they can't be used for regression. 

Saving as sales by month

```python
data_2023_24_filtered.to_csv('izev_data\\izev_tesla_purchases_2021-01_to_2024-03_2021-2024_izev_data.csv', index=False)
data_2019_23_filtered.to_csv('izev_data\\izev_tesla_purchases_2021-01_to_2024-03_2019-2023_izev_data.csv', index=False)
```

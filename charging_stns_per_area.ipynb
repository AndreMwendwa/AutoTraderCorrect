```python
import pandas as pd
import geopandas as gpd
```

```python
large_zones = gpd.read_file(
    r"C:\Users\mwendwa.kiko\Documents\Personal_Kiko\OneDrive\OneDrive - University of Toronto\UofT\Research\EV Adoption\Canada_Combined\Canada_Combined.shp"
)
```

```python
charging_stns = pd.read_excel(r'C:\Users\mwendwa.kiko\Documents\Personal_Kiko\OneDrive\OneDrive - University of Toronto\AutoTrader\Secondary Data\Charging Station Data Including Opening Date.xlsx')
charging_stns
```

```python
acceptable_facilities = ['GAS_STATION', 'RESTAURANT', 'PARKING_LOT', 'SHOPPING_CENTER', 
                         'GROCERY', 'UTILITY', 'FUEL_RESELLER', 'TRAVEL_CENTER', 
                         'SHOPPING_MALL', 'BREWERY_DISTILLERY_WINERY']
```

```python
charging_stns['Groups With Access Code'].value_counts()
```

```python
criteria = (charging_stns['Fuel Type Code'] == 'ELEC') & (charging_stns['Latitude'].notna() & (charging_stns['Facility Type'].isin(acceptable_facilities)))
filtered_charging_stns = charging_stns[criteria]
filtered_charging_stns
```

```python
charging_stns['Facility Type'].value_counts()
```

```python
# Convert to geodataframe
filtered_charging_stns_gdf = gpd.GeoDataFrame(
    filtered_charging_stns, geometry=gpd.points_from_xy(filtered_charging_stns['Longitude'], filtered_charging_stns['Latitude']))
# set crs as wgs84
filtered_charging_stns_gdf.set_crs(epsg=4326, inplace=True)
filtered_charging_stns_gdf = filtered_charging_stns_gdf.to_crs(large_zones.crs)
filtered_charging_stns_gdf
```

```python
# Merge with zones, counting charging stns in each zone
merged = gpd.sjoin(large_zones, filtered_charging_stns_gdf, how="left", predicate="intersects")
# charging_stns_count = merged.groupby("ZoneID").size()
charging_stns_count = merged.groupby("ZoneID")["EV Level2 EVSE Num"].sum()
large_zones["Charging_stations"] = large_zones["ZoneID"].map(charging_stns_count).fillna(0)
large_zones["Charging_stations"].describe()
```

```python
large_zones["Charging_stations"].sum()
```

```python
large_zones["Area"] = large_zones.geometry.area * 0.000001  # Convert to square kilometers
large_zones["Charging_stations_density"] = large_zones["Charging_stations"] / large_zones["Area"]
large_zones["Charging_stations_density"].describe()
```

```python
large_zones[["ZoneID", "Charging_stations", "Area", "Charging_stations_density"]].to_csv("charging_stations_per_area.csv", index=False)
```
